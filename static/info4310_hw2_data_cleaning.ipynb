```python
import numpy as np
import pandas as pd
import statistics
```

```python
df = pd.read_csv('fifa_unclean.csv')
```

```python
df.head()
```

```python
# group by team name, count each's frequency
counts = df.groupby(df['Team'])["Team"].count()
counts
```

```python
equalnum = counts[counts == 20]
equalnum
```

```python
for team in equalnum.keys():
  print(team)
```

```python
df20 = df.loc[df['Team'].isin(equalnum.keys())]
df20
```

```python
tmp = df20.loc[df20['Team'] == "Morecambe"]
tmp
```

```python
for team in equalnum.keys():
  print(team, ": Overall_Rating ranges from:   ", min(df20.loc[df20['Team'] == team]["Overall_Rating"]), max(df20.loc[df20['Team'] == team]["Overall_Rating"]), "   avg = ", statistics.mean(df20.loc[df20['Team'] == team]["Overall_Rating"]))
```

```python
# filter to just 5 teams that have good range of vals and are from around the globe (aka not all europe)
the5teams = ["Ayacucho", "Olympiakos Piraeus", "Bordeaux", "Belenenses", "Shenzhen"]

my5df = df20.loc[df20['Team'].isin(the5teams)]
my5df
```

```python
my5df.columns
```

```python
# convert value column to numbers rather than decimal millions (stil euros)
#my5df["Value € M"] = my5df["Value € M"] * 1000000.0
my5df["Value € M"] = my5df["Value € M"].apply(lambda x: x*1000000.0)
my5df
```

```python
my5df.rename(columns={'Value € M': 'Value_euros'}, inplace=True)
my5df
```

```python
my5df.to_csv('fifa_filtered.csv')
```
